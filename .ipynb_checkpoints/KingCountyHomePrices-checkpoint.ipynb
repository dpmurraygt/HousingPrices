```python
#PCA of the HousePrices to two principal components
#First create a pandas dataframe of the different square footage measures
X=pd.DataFrame(scale(HousePrices[['sqft_living', 'sqft_lot', 'sqft_living15', 'sqft_basement', 'sqft_above']]))

#Perform a PCA of the HousePrices Data
pca = PCA(n_components=5)
X_pca = pca.fit(X).transform(X) # fit data and then transform it
X_pca=pd.DataFrame(X_pca)
```

```python
Label=pd.DataFrame(['sqft_living', 'sqft_lot', 'sqft_living15', 'sqft_basement', 'sqft_above'])

Loadings=pd.DataFrame(pca.components_)
Loadings.insert(0,'Label', value=Label)
Loadings
```

A review of the loadings for the first principal compenent show the largest contributions are from the Square footage of the living space, followed by the square footage of the basement.  The above ground square footage of the home has a -0.7 loading, which would also likely be a large contribution.  The second principal component is dominated by the square footage of the living space at 0.91.  The third principal component has much smaller values for the loadings of the variables.

```python
#How much of the total variance of the six Square Footage statistics does our first 4 PC's explain?
print(sum(pca.explained_variance_ratio_[0:3]))
pca2=pd.DataFrame(pca.explained_variance_ratio_)
pca2=pca2.rename(columns = {0:'PC'})
print(pca2)
pca2.plot()
```

The first principal component on the six square footage statistics explains 54% of variance of the fields, the second explains 21.4% of the variance, and the third 18.6%, for a collective 94% of variance of these fields.  Based on inspection of the scree plot, the first three principal components can be used to represent the five variables.  The fit values for the first three principal components will be returned to the main data frame to represent the composite variability of the five square footage measures.

```python
X_pca=X_pca.rename_axis({0:'PC1', 1:'PC2', 2:'PC3', 3:'PC4', 4:'PC5'}, axis=1)
HousePrices=pd.concat([HousePrices, X_pca[['PC1', 'PC2', 'PC3']]], axis=1)
HousePrices.head(5)
```

```python
cls = KMeans(n_clusters=5, init='k-means++',random_state=1)

PriceKMeans = cls.fit_predict(HousePrices['price'].reshape(-1, 1))

```

```python
sns.pairplot(HousePrices[['bedrooms', 'bathrooms', 'PC1', 'PC2', 'floors', 'waterfront', 
                          'view', 'condition', 'yr_built', 'price'
        ]])
```

```python
#PriceKMeans as a DataFrame and then returned back into the HousePrices DataFrame
Price=HousePrices['price']
PriceKMeans=pd.DataFrame(PriceKMeans)
PriceKMeans2 = pd.concat([PriceKMeans[0:], Price], axis=1)
PriceKMeans2=PriceKMeans2.rename(columns = {0:'PriceCluster'})
PriceKMeans2.head(5)

HousePrices=pd.concat([HousePrices, PriceKMeans2], axis=1)
HousePrices.head(5)

#PriceKMeans2.groupby(price).describe
```
